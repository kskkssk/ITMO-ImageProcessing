```python
file_name = 'russia-2396022_1280.jpg'
img = cv2.imread(file_name)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
c = pywt.dwt2(img_gray, 'haar')
cA, (cH, cV, cD) = c
titles = ['cA', 'cH', 'cV', 'cD']
fig = plt.figure(figsize=(20, 4))
for i, a in enumerate([cA, cH, cV, cD]):
  ax = fig.add_subplot(1, 4, i + 1)
  ax.imshow(a, interpolation="nearest", cmap = 'gray')
  ax.set_title(titles[i], fontsize = 20)
  fig.tight_layout()
plt.show()
```

```python
#Building a histogram of the cH set
grayscale_hist = plt.hist(cH.ravel(), bins = 256)
plt.xlabel('Intensity $n$')
plt.ylabel('Number $h(n)$')
plt.show()
```

```python
print('The maximum value of the function h(n):', max(grayscale_hist[0]))
```
